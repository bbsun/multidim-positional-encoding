```python
import torch
from positional_encodings.torch_encodings import PositionalEncoding1D, PositionalEncoding2D, PositionalEncoding3D, PositionalEncoding4D,Summer
import numpy as np
import matplotlib.pyplot as plt
p_enc_2d = PositionalEncoding3D(256)
y = torch.zeros((1,10,10,10,256))
print(p_enc_2d(y).shape) # (1, 6, 2, 8)
y = p_enc_2d(y).reshape(-1,256)
plt.figure(dpi=200)
plt.imshow(np.transpose(y.detach().cpu().numpy()),cmap='gray',aspect=1)
plt.show()

# plt.figure(dpi=1200)
# plt.imshow(np.transpose(y.detach().cpu().numpy()),cmap='gray',aspect=2)
# plt.show()
p_enc_4d = PositionalEncoding4D(1024)
y = torch.zeros((1,5,5,5,5,1024))
y= p_enc_4d(y)
y=y.reshape(-1,1024)
plt.figure(dpi=200)
plt.imshow(np.transpose(y.detach().cpu().numpy()),cmap='gray',aspect=1)
plt.show()
```
